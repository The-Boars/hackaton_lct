```python
import ast
import pandas as pd
from collections import Counter

df = pd.read_csv('../datasets/submission.csv', sep=';')

def extract_labels(span_text):
    if pd.isna(span_text) or not str(span_text).strip():
        return []
    try:
        spans = ast.literal_eval(span_text)
    except (ValueError, SyntaxError):
        return []
    return [span[2] for span in spans if len(span) >= 3]

label_counter = Counter()
for labels in df['annotation'].map(extract_labels):
    label_counter.update(labels)

class_balance = pd.DataFrame(
    label_counter.items(), columns=['label', 'count']
).sort_values('count', ascending=False)

display(class_balance)
print(f'Total labeled spans: {sum(label_counter.values())}')
```

```python
import ast
import pandas as pd

pd.options.display.max_colwidth = None
pd.options.display.max_rows = 700

# файл без header -> читаем с header=None, разделитель таб
submission = pd.read_csv('../datasets/valid.csv', sep='\t', header=None)

annotation_col = submission.columns[-1]  # последний столбец содержит аннотацию

def has_brand_tag(annotation_str):
    if not isinstance(annotation_str, str):
        return False
    tuples = ast.literal_eval(annotation_str)
    return any('I-BRAND' in tag for _, _, tag in tuples)

rows_with_brand = submission[submission[annotation_col].apply(has_brand_tag)]
rows_with_brand

```

```python
import ast
import pandas as pd

pd.options.display.max_colwidth = None
pd.options.display.max_rows = 700

submission = pd.read_csv('../datasets/submission.csv', sep=';')

def has_label_o(annotation_str):
    tuples = ast.literal_eval(annotation_str)
    return any(item[2] == 'O' for item in tuples)

rows_with_label_o = submission[submission['annotation'].apply(has_label_o)]
rows_with_label_o
```
