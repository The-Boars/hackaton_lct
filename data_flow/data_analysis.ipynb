```python
import pandas as pd
import numpy as np
```

```python
train_data = pd.read_csv('./Датасет/train.csv', sep=';')
```

```python
first_col = train_data.columns[0]
word_counts = (
    train_data[first_col]
    .str.lower()
    .str.split(r"\s+")      # разделяем по пробелам/табам
    .explode()              # разворачиваем список слов в строки
    .value_counts()
)

print(word_counts.head(1000).to_string())  # выводит топ-1000 слов
```
